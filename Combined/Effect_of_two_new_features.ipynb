```python
import pandas as pd
import numpy as np
from scipy.stats import ttest_ind,wilcoxon,mannwhitneyu
from IPython.display import display,clear_output,HTML
```

```python
df_test=pd.read_csv('../Data/Test_Data/df_fin.csv',index_col='domain')
df_test_url=pd.read_csv('../Data/Test_Data/df_url.gz',index_col='domain',compression='gzip')
df=pd.concat([df_test,df_test_url[['url_char_-','url_number_special_chars']]],join='inner',axis=1)
```

### Both tests are for dataset2 only

### For 3% of Alexa1m data 

#### Correlation values for all independent variables.

```python
n=0.03
# print("For {}% of Alexa1m data \n".format(int(n*100)))
df_frac=df.sample(frac=1,random_state=100)
df=df_frac.iloc[:int(df_frac.shape[0]*n),:]
```

```python
lst=[]
for col in df.columns:
    if col!='Target':
        lst.append((col,df[col].corr(df['Target'])))
df_corr=pd.DataFrame(lst)
df_corr.columns=['Feature','Correlation']
df_corr.Correlation=df_corr.Correlation.apply(lambda x: np.abs(x))
df_corr.sort_values(by='Correlation',ascending=False,inplace=True)
df_corr
```

#### Wilcoxon-Mann-Whitney test for discrete variables

```python
lst=[]
for col in df.columns:
    if col!='Target' and 'http' not in col:
        wil_stats=mannwhitneyu( df[col][df['Target']==0] , df[col][df['Target']==1],alternative='two-sided')
        wil_val=wil_stats[0]
        pval=wil_stats[1]
        lst.append( ( col, wil_val , pval ) )

df_ttest=pd.DataFrame(lst)
df_ttest.columns=['Feature','Wilcoxon_Mann_Whitney_U_stat','p-val']
df_ttest.Wilcoxon_Mann_Whitney_U_stat=df_ttest.Wilcoxon_Mann_Whitney_U_stat.apply(lambda x: np.abs(x))
df_ttest.sort_values(by='Wilcoxon_Mann_Whitney_U_stat',ascending=True,inplace=True)
df_ttest
```
